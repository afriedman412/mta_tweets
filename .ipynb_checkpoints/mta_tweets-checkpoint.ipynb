```python
tw = '@johannagriese Good Afternoon. Can you tell us your location where you are waiting to board this train so we may investigate? ^JZ'
```

```python
'@' in tw
```

```python
def cleany(tweet):
    tweet = re.sub('@\S+','', tweet)
    tweet = re.sub('\^\S+','', tweet)
    tweet = re.sub('https:\S+','', tweet)
    return tweet.lower()
```

```python
cleany(tw)
```

```python
tw2 = '@MTAinWork @NYPDTransit We have alerted supervision to locate this train. We apologize for your unpleasant commute this evening. We are in the process of addressing this immediately. ^JZ'
```

```python
cleany(tw2)
```

```python
df['tweet_clean'] = df.text.map(cleany)
```

```python
def find_sig(tw):
    if re.search('\^\S{1,4}', tw) != None:
        return re.search('\^\S{1,4}', tw)[0]
    else:
        return None
```

```python
df.sig = df.text.map(find_sig)
```

```python
data = df[df.subjectivity > 0]
a = 0.3

plt.figure(figsize=(15,6))
plt.subplot(131)
data.subjectivity.hist(bins=50, color='Orange', ec='white', lw=2)
plt.title(s='subjectivity distribution')

plt.subplot(132)
data.polarity.hist(bins=50, color='Red', ec='white', lw=2)
plt.title(s='polarity distribution distribution')

plt.subplot(133)
plt.scatter(data=data, x='subjectivity' , y='polarity', alpha=a, s=3)
plt.title(s='subjectivity v. polarity');
```

```python
n = 1
plt.figure(figsize=(15,20))
plt.subplots_adjust(left=None, bottom=None, right=None, top=1.4,
                wspace=None, hspace=None)
for s in df.sig.unique():
    plt.subplot(7,4,n)
    data = df[df.sig == s]
    plt.scatter(data=data, x='subjectivity' , y='polarity', alpha=0.5, s=3)
    plt.title(s=s)
    n += 1;
```
